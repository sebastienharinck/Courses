# Matplotlib Features
The goal of this jupyter notebook is to have a view about basics matplotlib features.

## Imports

```python
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
from matplotlib import gridspec
import random
import numpy as np
```

## Matplotlib image
### Read image

```python
img = mpimg.imread('http://kordouane.k.o.pic.centerblog.net/f987af30.png')
```

```python
img.shape
```

### Display image

```python
plt.imshow(img)
plt.show()
```

### Generate image
I want to generate an image 28*28 px. So I generate a numpy array of 784 rows.

```python
img = np.array([random.choice([0, 1]) for _ in range(784)])
img = img.reshape(28, 28) # warning : not reshape the actual array
img.shape
```

Why it is (784,) ? I thought that it would be (784, 1) ...

```python
plt.imshow(img)
plt.show()
```

We can specify some colors and styles :

```python
plt.imshow(img, cmap='Greys')
plt.show()
```

## Diplay multiples images
I want to display 100 images samples from the MNIST dataset. I will try it with the GridSpec

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)
```

```python
samples = mnist.train.images[0:100]
samples.shape
```

### Try to display one image

```python
img = samples[1].reshape(28, 28)
plt.imshow(img, cmap='Greys')
plt.show()
```

### Try to display 100 images

```python
fig = plt.figure(figsize=(10,10))
gs = gridspec.GridSpec(10,10)

for i, image in enumerate(samples):
    plt.subplot(gs[i])
    plt.axis('off')
    plt.imshow(image.reshape(28,28), cmap='Greys')
```

```python
plt.show()
```

## More information on an image

```python
image, label = mnist.train.images[11].reshape(28,28), np.argmax(mnist.train.labels[11])
plt.imshow(image, cmap='Greys')
plt.xlabel('True : {}'.format(label))
plt.show()
```
